```python
df3 = df3.drop(columns = ['hadm_id', 'itemid', 'icustay_id'])
```

```python
df4 = pd.read_csv('RR_clean.csv')
```

```python
df4 = df4.drop(columns = ['hadm_id', 'itemid', 'icustay_id'])
```

```python
df4
```

```python

# Assuming df1, df2, df3, df4 are your data frames

# Performing outer join with custom suffixes
result = df1.merge(df2, on=['subject_id', 'hourly_bin'], how='outer', suffixes=('_df1', '_df2'))
result = result.merge(df3, on=['subject_id', 'hourly_bin'], how='outer', suffixes=('', '_df3'))
result = result.merge(df4, on=['subject_id', 'hourly_bin'], how='outer', suffixes=('', '_df4'))
```

```python
result = result.drop(columns = ['row_id'])[['subject_id', 'hourly_bin', 'temp', 'WBC', 'HeartRate', 'RR']]
```

```python
result.to_csv('/Users/asifmahdin/Downloads/Clean_Result.csv', index=False)
```
